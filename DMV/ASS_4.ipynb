### Step 1: Importing Libraries and Loading the Dataset

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("RealEstate_Prices.csv")
```

### Step 2: Handling Missing Values

```python
df.isnull().sum()
```

```python
df['bedrooms']=df['bedrooms'].fillna(df['bedrooms'].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
df['bedrooms']
```

```python
df['area sq ft']=df['area sq ft'].fillna(df['area sq ft'].mean())
```

```python
df.isnull().sum()
```

```python
df['sale price']=df['sale price'].fillna(df['sale price'].median())
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.dtypes
```

```python
df[['bedrooms','area sq ft','sale price']]=df[['bedrooms','area sq ft','sale price']].astype('int64')
```

```python
df
```

### Step 3: Filtering and Subsetting the Data

```python
filtered_df = df[(df['neighborhood'] == 'Downtown') & (df['property type'] == 'Apartment')]
```

```python
filtered_df
```

### Step 4: Encoding Categorical Variables

```python
df['property type'].value_counts()
```

```python
df['property type']=df['property type'].replace({'House':1, 'Apartment':2, 'Condo':3})
```

```python
df
```

### Step 5: Aggregating Data

```python
aggregated_df = df.groupby("neighborhood").agg({"sale price": "mean", "area sq ft": "mean"})
aggregated_df
```

### Step 6: Identifying and Handling Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(df)
```

```python
features = ['area sq ft', 'sale price']

for feature in features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df)
```
