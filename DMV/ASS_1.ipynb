```python
import pandas as pd

# Load CSV file
sales_csv = pd.read_csv("customers.csv")

# Load Excel file
sales_excel = pd.read_excel("customers.xlsx")

# Load JSON file
sales_json = pd.read_json("customers.json")
```

### 2. Explore the Structure and Content of Each Dataset

```python
# Check for missing values, data types, and structure in each dataset
print("CSV Data Info:\n", sales_csv.info())
print("Excel Data Info:\n", sales_excel.info())
print("JSON Data Info:\n", sales_json.info())

# Check for any inconsistencies or issues with data quality
print("CSV Missing Values:\n", sales_csv.isnull().sum())
print("Excel Missing Values:\n", sales_excel.isnull().sum())
print("JSON Missing Values:\n", sales_json.isnull().sum())
```

### 3. Perform Data Cleaning

```python
# Remove duplicates from each dataset
sales_csv.drop_duplicates(inplace=True)
sales_excel.drop_duplicates(inplace=True)
sales_json.drop_duplicates(inplace=True)

# Fill missing values (using mean or mode as appropriate)
sales_csv.fillna(0, inplace=True)
sales_excel.fillna(sales_excel.mode(), inplace=True)
sales_json.fillna(sales_json.mode(), inplace=True)
```

### 4. Convert Data into a Unified Format

```python
# Combine all dataframes into a single one
combined_sales_data = pd.concat([sales_csv, sales_excel, sales_json], ignore_index=True)
```

```python
combined_sales_data.head()
```

```python
combined_sales_data.shape
```

### 5. Data Transformation

Merging first_name and last_name and create new column='full_name'

```python
# Example: Creating a new column for total sales by multiplying quantity by unit price
combined_sales_data['full_name'] = combined_sales_data['first_name'] + ' ' + combined_sales_data['last_name']
```

```python
combined_sales_data.head()
```

splitting columns

```python
# Example: Splitting 'Order Date' into 'Year' and 'Month'
combined_sales_data['Year'] = pd.DatetimeIndex(combined_sales_data['registered']).year
combined_sales_data['Month'] = pd.DatetimeIndex(combined_sales_data['registered']).month
```

```python
combined_sales_data.head()
```

### 6. Analyze the Data with Descriptive Statistics

```python
# Total orders
total_orders = combined_sales_data['orders'].sum()
print("Total Orders:", total_orders)
```

```python
combined_sales_data.groupby('job').agg({'age':'mean','orders':'sum','spent':'mean'})
```

```python
combined_sales_data['job'].value_counts()
```

### 7. Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='orders', data=combined_sales_data)
plt.title("Monthly Sales Distribution")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Bar plot of Order value distribution
plt.figure(figsize=(15,6))
sns.boxplot(x='job', y='spent', data=combined_sales_data)
plt.title('Order value Distribution by job')
plt.xlabel('JOB')
plt.ylabel('Order Value')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart of sales by category
plt.figure(figsize=(10, 10))
combined_sales_data.groupby('job')['orders'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Orders by job category")
plt.ylabel("")
plt.show()
```
