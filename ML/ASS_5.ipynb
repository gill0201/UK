## Ensemble Learning
A. Implement Random Forest Classifier model to predict the safety of the car. 
Dataset link: https://www.kaggle.com/datasets/elikplim/car-evaluation-data-set 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('car.csv')
```

```python
df
```

```python
df.columns=['buing','maint','doors','persons','lug_boot','class','safety']
df.head(10)
```

```python
oe=OrdinalEncoder()
df[['buing','maint','doors','persons','lug_boot','class','safety']]=oe.fit_transform(df[['buing','maint','doors','persons','lug_boot','class','safety']])
df.head()
```

```python
x=df.drop(columns=['safety'])
y=df['safety']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train,y_train)
```

```python
y_pred=rf_model.predict(X_test)
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
